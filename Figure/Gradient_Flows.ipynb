```python
gaussian_SW = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d100_SW.txt")
gaussian_maxSW = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d100_maxSW.txt")
gaussian_minSW_mc = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d100_minSW_mc.txt")
gaussian_minSW_opt = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d100_minSW_opt.txt")
res_gaussians_SW = np.mean(gaussian_SW,axis=0)
res_gaussians_maxSW = np.mean(gaussian_maxSW,axis=0)
res_gaussians_minSW_mc = np.mean(gaussian_minSW_mc,axis=0)
res_gaussians_minSW_opt = np.mean(gaussian_minSW_opt,axis=0)

gaussian_SW8 = np.loadtxt("../Notebook/Results/Gradient_Flow/8gaussians_SW.txt")
gaussian_maxSW8 = np.loadtxt("../Notebook/Results/Gradient_Flow/8gaussians_maxSW.txt")
gaussian_minSW_mc8 = np.loadtxt("../Notebook/Results/Gradient_Flow/8gaussians_minSW_mc.txt")
gaussian_minSW_opt8 = np.loadtxt("../Notebook/Results/Gradient_Flow/8gaussians_minSW_opt.txt")
res_8gaussians_SW = np.mean(gaussian_SW8,axis=0)
res_8gaussians_maxSW = np.mean(gaussian_maxSW8,axis=0)
res_8gaussians_minSW_mc = np.mean(gaussian_minSW_mc8,axis=0)
res_8gaussians_minSW_opt = np.mean(gaussian_minSW_opt8,axis=0)

gaussian_SW25 = np.loadtxt("../Notebook/Results/Gradient_Flow/25gaussians_SW.txt")
gaussian_maxSW25 = np.loadtxt("../Notebook/Results/Gradient_Flow/25gaussians_maxSW.txt")
gaussian_minSW_mc25 = np.loadtxt("../Notebook/Results/Gradient_Flow/25gaussians_minSW_mc.txt")
gaussian_minSW_opt25 = np.loadtxt("../Notebook/Results/Gradient_Flow/25gaussians_minSW_opt.txt")
res_25gaussians_SW = np.mean(gaussian_SW25,axis=0)
res_25gaussians_maxSW = np.mean(gaussian_maxSW25,axis=0)
res_25gaussians_minSW_mc = np.mean(gaussian_minSW_mc25,axis=0)
res_25gaussians_minSW_opt = np.mean(gaussian_minSW_opt25,axis=0)


halfmoon_SW = np.loadtxt("../Notebook/Results/Gradient_Flow/half_moons_SW.txt")
halfmoon_maxSW = np.loadtxt("../Notebook/Results/Gradient_Flow/half_moons_maxSW.txt")
halfmoon_minSW_mc = np.loadtxt("../Notebook/Results/Gradient_Flow/half_moons_minSW_mc.txt")
halfmoon_minSW_opt = np.loadtxt("../Notebook/Results/Gradient_Flow/half_moons_minSW_opt.txt")
res_halfmoon_SW = np.mean(halfmoon_SW,axis=0)
res_halfmoon_maxSW = np.mean(halfmoon_maxSW,axis=0)
res_halfmoon_minSW_mc = np.mean(halfmoon_minSW_mc,axis=0)
res_halfmoon_minSW_opt = np.mean(halfmoon_minSW_opt,axis=0)

gaussian_SW500d = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_500d_SW.txt")
gaussian_maxSW500d = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_500d_maxSW.txt")
gaussian_minSW_mc500d = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_500d_minSW_mc.txt")
gaussian_minSW_opt500d = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_500d_minSW_opt.txt")
res_500dgaussians_SW = np.mean(gaussian_SW500d,axis=0)
res_500dgaussians_maxSW = np.mean(gaussian_maxSW500d,axis=0)
res_500dgaussians_minSW_mc = np.mean(gaussian_minSW_mc500d,axis=0)
res_500dgaussians_minSW_opt = np.mean(gaussian_minSW_opt500d,axis=0)

gaussian_SW1000 = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d_SW1000.txt")
gaussian_maxSW1000 = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d_maxSW1000.txt")
gaussian_minSW_mc1000 = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d_minSW_mc1000.txt")
gaussian_minSW_opt1000 = np.loadtxt("../Notebook/Results/Gradient_Flow/gaussian_2d_minSW_opt1000.txt")
res_gaussians1000_SW = np.mean(gaussian_SW1000,axis=0)
res_gaussians1000_maxSW = np.mean(gaussian_maxSW1000,axis=0)
res_gaussians1000_minSW_mc = np.mean(gaussian_minSW_mc1000,axis=0)
res_gaussians1000_minSW_opt = np.mean(gaussian_minSW_opt1000,axis=0)
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})
plt.rcParams["figure.figsize"] = (6,13)

s = 6
fig = plt.figure(figsize=(6,12))
grid = plt.GridSpec(6, 2, wspace=.1, hspace=0.3, width_ratios=[3, 1.2])
titles = ["SW", "max-SW", "min-SWGG (Monte Carlo)", "min-SWGG (optim)"]

#gaussians
plt.subplot(grid[0, 0])
plt.plot(np.log10(res_gaussians_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_gaussians_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_gaussians_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_gaussians_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.legend(titles, fontsize=10, bbox_to_anchor=(0, 1.4, 1, 0.2), loc="upper left", mode="expand", ncol=2)
#plt.title('2-Wasserstein Distance', fontsize=10)
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[0, 1])
plt.scatter(X_gaussian[:,0], X_gaussian[:,1], marker='x')
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)

#8gaussians
plt.subplot(grid[1, 0])
plt.plot(np.log10(res_8gaussians_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_8gaussians_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_8gaussians_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_8gaussians_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[1, 1])
plt.scatter(X_8gaussian[:,0], X_8gaussian[:,1], marker='x')
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)

#25gaussians
plt.subplot(grid[2, 0])
plt.plot(np.log10(res_25gaussians_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_25gaussians_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_25gaussians_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_25gaussians_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[2, 1])
plt.scatter(X_25gaussian[:,0], X_25gaussian[:,1], marker='x')
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)

#halfmoon
plt.subplot(grid[3, 0])
plt.plot(np.log10(res_halfmoon_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_halfmoon_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_halfmoon_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_halfmoon_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[3, 1])
plt.scatter(X_halfmoon[:,0], X_halfmoon[:,1], marker='x')
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)


#500dgaussians
plt.subplot(grid[4, 0])
plt.plot(np.log10(res_500dgaussians_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_500dgaussians_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_500dgaussians_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_500dgaussians_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[4, 1])
plt.scatter([], [])
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)
plt.text(-0.04, 0, "Gaussian 500d")

#1000gaussians
plt.subplot(grid[5, 0])
plt.plot(np.log10(res_gaussians1000_SW), "--",color="C4", linewidth=1,  label = "SW")
plt.plot(np.log10(res_gaussians1000_maxSW), "--",color="C3",linewidth=1,  label = "max-SW")
plt.plot(np.log10(res_gaussians1000_minSW_mc), color="C0",linewidth=1,  label = "min-SWGG (Monte Carlo)")
plt.plot(np.log10(res_gaussians1000_minSW_opt),color="C1",linewidth=1,  label = "min-SWGG (optim)")
plt.ylabel(r'$Log_{10}(W_2)$')

plt.subplot(grid[5, 1])
plt.scatter(X_gaussian1000[:,0], X_gaussian1000[:,1], marker='x')
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)


plt.tight_layout()
#plt.savefig("GF.pdf",bbox_inches='tight')
```
