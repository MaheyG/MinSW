```python
import sys
path='/minS/'
sys.path.append(path)

import numpy as np
import torch
import matplotlib.pylab as pl
import ot
import sklearn
import pylab
from tqdm import trange

from generate_data import make_data
from minS_Optim import get_minSW_smooth,upperW2_smooth
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

device='cpu'
dtype = torch.DoubleTensor
print('Used device: '+device)
```

```python
n=1000
dim=2
xs,xt=make_data('gaussians',n,dim)


X = torch.from_numpy(xs).to(device)
Y = torch.from_numpy(xt).to(device)
X=X.float()
Y=Y.float()

pl.figure(figsize=(3,3))
pl.scatter(xs[:, 0], xs[:, 1], c='C0', label='Source')
pl.scatter(xt[:, 0], xt[:, 1], c='C1', label='Target')
pl.legend(loc=2)
```

```python
num_iter=200
lr=1e0
s=20
std=.5

t,loss_smooth_l,proj_smooth_l = get_minSW_smooth(X,Y,lr=lr,num_iter=num_iter,s=s,std=std)
t.requires_grad=False
theta_optim=t/torch.norm(t)

res_smooth=upperW2_smooth(X,Y,theta_optim,s=1,std=0)
print('minS with selected theta',res_smooth)
```

```python
pl.figure(figsize=(8,3))
grid = pl.GridSpec(1, 2, wspace=0.4, hspace=0.3)

pl.subplot(grid[0, 0])
pl.plot(loss_smooth_l)
pl.title('minSW smooth mass')

pl.subplot(grid[0, 1])
pl.scatter(X[:, 0], X[:, 1], c='C0', label='Source')
pl.scatter(Y[:, 0], Y[:, 1], c='C1', label='Target')
for i in range(num_iter):
    pl.axline((0,0), list(proj_smooth_l[i,0:2]), color='C2',alpha=(i/num_iter)**1)
```
